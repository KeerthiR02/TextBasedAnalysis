**Regex**

**Function to tokenize:**

```python
import regex as re
def tokenize(text):
  return re.findall(r'[\w-]*[a-z|A-Z][\w-]*', text)
```

```python
testtext = "NLTK is a leading platform for building Python programs to work with human language data."
testtokens = tokenize(testtext)
print(testtokens)
```

**How to Remove stop words**

```python
import nltk
nltk.download('stopwords')
stopwords = set(nltk.corpus.stopwords.words('english'))
print(stopwords)
```

**Function to remove stop words**

```python
def remove_stop(tokens):
  return [t for t in tokens if t.lower() not in stopwords]
  cleantokens= remove_stop(testtokens)
  print(cleantokens)
```

```python
pipeline = [str.lower, tokenize, remove_stop]
```

**Let’s make a “pipeline”**

```python
def prepare(text, pipeline):
  tokens = text
  for fun in pipeline:
    tokens = fun(tokens)
  return tokens
```

```python
import pandas as pd

# Assuming df is your DataFrame and 'college' is the column containing text data
# Convert the 'college' column to string type
df['state'] = df['state'].astype(str)

# Now apply the tokenization function
df['tokens'] = df['state'].apply(prepare, pipeline=pipeline)
```

**Count the tokens**

```python
df['num_tokens'] = df['tokens'].map(len)
df.tail()
```

**Word frequencyAnalysis**

```python
from collections import Counter

text = "The quick brown fox jumps over the lazy dog."
tokens = text.split()  # Split the text into tokens (words)

word_count = Counter(tokens)
print(word_count)
```

**You can also add new data to the currentcount:**


more_tokens = tokenize("The quick brown fox jumps over the lazy dog.")
word_count.update(more_tokens)
print(word_count)

**Let’s count:**

```python
counter = Counter()
df['tokens'].map(counter.update)
print(counter.most_common(5))
```

**Now, we put the word frequency in it’s owndataframe for analysis**

```python
freq_df = pd.DataFrame.from_dict(counter,orient='index', columns=['freq'])
freq_df = freq_df.query('freq >= 1')
freq_df.index.name = 'token'
```

```python
freq_df = freq_df.sort_values('freq', ascending=False)
freq_df.head(10)
```

**Word cloud**

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

# Sample text
text = "Data science is an interdisciplinary field that uses scientific methods, algorithms, processes, and systems to extract knowledge and insights from structured and unstructured data."

# Define stopwords
stopwords = set(STOPWORDS)

# Create a WordCloud object
wc = WordCloud(max_words=100, stopwords=stopwords, background_color="white")

# Generate the word cloud
wc.generate(text)

# Plot the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```
