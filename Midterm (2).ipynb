```python
import pandas as pd
#Filepath
file = "C:\\Users\\keert\\Downloads\\collection.csv"
df = pd.read_csv(file)

df.head(20)
```

```python
import nltk
nltk.download('stopwords')
stopwords = set(nltk.corpus.stopwords.words('english'))
print(stopwords)


```

```python
import regex as re
def tokenize(text):
    return re.findall(r'[\w,:.-]*[a-zA-Z][\w,:.-]*', text)
def remove_stop(tokens):
    return [t for t in tokens if t.lower() not in stopwords]
pipeline = [str.lower, tokenize, remove_stop]
def prepare(text, pipeline):
  tokens = text
  for fun in pipeline:
    tokens = fun(tokens)
  return tokens
df['tokens'] = df['text'].apply(prepare, pipeline=pipeline)
```

```python

import pandas as pd
from collections import Counter

counter = Counter()
df['tokens'].map(counter.update)
freq_df = pd.DataFrame.from_dict(counter, orient='index', columns=['freq'])
freq_df = freq_df.query('freq >= 1')
freq_df.index.name = 'tokens'
freq_df = freq_df.sort_values('freq', ascending=False)
freq_df.head(10)
```

```python
#QUESTION 4

import re

# Sample strings
strings = [
    "Username: john_doe123",
    "Password: P@ssw0rd",
    "Date of Birth: 1990-01-01",
    "Credit Card: 1234-5678-9012-3456",
    "SSN: 123-45-6789"
]

# Regex patterns
patterns = {
    "Username": r'\bUsername:\s*([a-zA-Z0-9_]+)\b',
    "Password": r'\bPassword:\s*(\S+)\b',
    "Date": r'\bDate of Birth:\s*(\d{4}-\d{2}-\d{2})\b',
    "Credit Card": r'\bCredit Card:\s*((?:\d{4}[-\s]?){3}\d{4})\b',
    "SSN": r'\bSSN:\s*(\d{3}-\d{2}-\d{4})\b'
}

# Dictionary to store matched strings
sorted_strings = {pattern: [] for pattern in patterns}

# Matching strings with regular expressions
for string in strings:
    for pattern, regex in patterns.items():
        match = re.search(regex, string)
        if match:
            sorted_strings[pattern].append(match.group(1))

# Printing matched strings
for pattern, matched_strings in sorted_strings.items():
    print(f"{pattern}:")
    for string in matched_strings:
        print(f"  {string}")

```

```python

import re

# Sample strings containing the patterns
username_example = "Usernames can contain letters, numbers, and underscores (e.g., john_doe123)."
password_example = "Passwords must be at least 8 characters long and contain at least one uppercase letter, one lowercase letter, one digit, and one special character (!@#$%^&*)."
date_example = "The report was published on 12/25/2021 and updated on 01/05/2022. The meeting is scheduled for 03/15/2023."
credit_card_example = "You can pay using your credit card: 1234-5678-9012-3456"
ssn_example = "Please provide your SSN: 987654321."
# Regex patterns
username_pattern = r'\b[a-zA-Z0-9_]{4,}\b'
password_pattern = r'^(?=.*[a-z])(?=.*[A-Z])(?=.*\d)(?=.*[!@#$%^&*])[A-Za-z\d!@#$%^&*]{8,}$'
date_pattern = r'\b\d{2}/\d{2}/\d{4}\b'
credit_card_pattern = r'\b(?:\d{4}[-\s]?){3}\d{4}\b'
ssn_pattern = r'\b\d{3}-\d{2}-\d{4}\b'
# Extracting and printing matches
print("Username:", re.findall(username_pattern, username_example))
print("Password:", re.findall(password_pattern, password_example))
print("Date:", re.findall(date_pattern, date_example))
print("Credit Card:", re.findall(credit_card_pattern, credit_card_example))
print("SSN Example :", re.findall(ssn_pattern, ssn_example))
```

```python
#QUESTION 5

import pandas as pd
from textblob import TextBlob

# Sample dataset
data = {
    'AUTHOR': ['Lisa Wellas', 'jason graham', 'Ajkal Khan', 'Dakota Taylor'],
    
}

df = pd.DataFrame(data)

# Perform sentiment analysis
def analyze_sentiment(text):
    blob = TextBlob(text)
    sentiment_score = blob.sentiment.polarity
    if sentiment_score > 0:
        return 'Positive'
    elif sentiment_score < 0:
        return 'Negative'
    else:
        return 'Neutral'



df['sentiment'] = df['AUTHOR'].apply(analyze_sentiment)


# Display the results
print(df)
```

```python
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt

# Sample dataset
# Load your dataset
file="C:\\Users\\keert\\Downloads\\collection.csv"
df=pd.read_csv(file)# Replace "your_dataset.csv" with the path to your dataset



# Perform sentiment analysis
def analyze_sentiment(text):
    blob = TextBlob(text)
    sentiment_score = blob.sentiment.polarity
    if sentiment_score > 0:
        return 'Positive'
    elif sentiment_score < 0:
        return 'Negative'
    else:
        return 'Neutral'

df['sentiment'] = df['AUTHOR'].apply(analyze_sentiment)





# Plot results
sentiment_counts = df['sentiment'].value_counts()
# Plot the distribution of sentiments
plt.figure(figsize=(8, 6))
sentiment_counts.plot(kind='bar', color=['green', 'red', 'blue'])
plt.title('Distribution of Sentiments in Player Names')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```
